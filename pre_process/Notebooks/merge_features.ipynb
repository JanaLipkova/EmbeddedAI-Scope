```python
import pandas as pd
import h5py
import numpy as np
from os.path import join as j
import os
import torch
```

```python
feat_dir = '/media/fedshyvana/ssd1/resnet_features/bwh_lung_microscopy_crop_resnet_features'
#feat_dir = '/media/fedshyvana/ssd1/ghostnet_features/bwh_lung_microscopy_crop_ghostnet_features'
#feat_dir = '/media/fedshyvana/ssd1/mobilenet_features/bwh_lung_microscopy_crop_mobilenet_features'
```

```python
pt_dir = j(feat_dir, 'pt_files')
sink_dir = j(feat_dir + '_merged', 'pt_files')
os.makedirs(sink_dir, exist_ok=True)
files = sorted(os.listdir(pt_dir))
```

```python
df = pd.DataFrame(files, columns=['img_name'])
df['slide_id'] = df['img_name'].str.split('_').str[:2].str.join('_')
df
```

```python
for slide in df.slide_id.unique():
    pt_files = df[df.slide_id == slide]['img_name'].values
    print(pt_files)
    features = [torch.load(j(pt_dir, pt_file)) for pt_file in pt_files]
    features = torch.cat(features, dim=0)
    torch.save(features, os.path.join(sink_dir, slide+'.pt'))
    print(features.shape)
```
